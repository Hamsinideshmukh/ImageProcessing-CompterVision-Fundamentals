```python
def sobel(img):
    imgx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
    imgy = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)

    res = cv2.magnitude(imgx, imgy)

    return res

def prewitt(img):
    kernalPrewittx = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
    kernalPrewitty = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])

    imgx = cv2.filter2D(img, -1, kernalPrewittx)
    imgy = cv2.filter2D(img, -1, kernalPrewitty)

    res = cv2.magnitude(imgx.astype(np.float32), imgy.astype(np.float32))

    return res

e = sobel(img)

# Set the scaling constant
c = 0.3

# Calculate the sharpened image
g = img + c * e

# Clip the values to be in the valid range [0, 255]
g = np.clip(g, 0, 255)

# Convert to an 8-bit unsigned integer type
g = g.astype(np.uint8)

fig, axes = plt.subplots(1, 2, figsize=(6, 3))
axes[0].imshow(img, cmap="gray")
axes[0].set_title("Original Image")
axes[0].set_axis_off()
axes[1].imshow(g, cmap="gray")
axes[1].set_title("Sharpened Image")
axes[1].set_axis_off()
```

```python
#Writing custom function for canny edge detection
def canny(img, t0, t1):
    #Gaussian Smoothing
    gaussian_img = cv2.GaussianBlur(img, (5, 5), 0)

    gx = cv2.Sobel(gaussian_img, cv2.CV_64F, 1, 0, ksize=3)
    gy = cv2.Sobel(gaussian_img, cv2.CV_64F, 0, 1, ksize=3)

    #Non-maximal suppression
    mag = cv2.magnitude(gx, gy)
    angle = cv2.phase(gx, gy)

    #Non-maxima supression
    nms = np.zeros_like(mag)
    for i in range(1, mag.shape[0] - 1):
        for j in range(1, mag.shape[1] - 1):
            #Get the gradient direction
            dir = angle[i, j] * 180 / np.pi  # convert to degrees
            dir = dir % 180
            q = 255
            r = 255
            if (0 <= dir < 22.5) or (157.5 <= dir <= 180):
                q = mag[i, j+1]
                r = mag[i, j-1]
            elif (22.5 <= dir < 67.5):
                q = mag[i+1, j-1]
                r = mag[i-1, j+1]
            elif (67.5 <= dir < 112.5):
                q = mag[i+1, j]
                r = mag[i-1, j]
            else:
                q = mag[i-1, j-1]
                r = mag[i+1, j+1]

            if(mag[i, j] >= q and mag[i, j] >=r):
                nms[i, j] = mag[i, j]
            else:
                nms[i, j] = 0

    #Hysteresis thresholding
    for i in range(1, nms.shape[0] - 1):
        for j in range(1, nms.shape[1] - 1):
            if(nms[i, j] < t0):
                nms[i, j] = 0
            elif t0 <= nms[i, j] < t1:
                if np.max(nms[i-1:i+2, j-1:j+2]) >= t1:
                    nms[i, j] = 255
                else:
                    nms[i, j] = 0
            else:
                nms[i, j] = 255

    return nms.astype(np.uint8)

canny_edges_custom = canny(img, 5, 105)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img, cmap="gray")
axes[0].set_title("Original Image")
axes[0].set_axis_off()
axes[1].imshow(canny_edges_custom, cmap="gray")
axes[1].set_title("Canny Edge Detection")
axes[1].set_axis_off()
plt.tight_layout()
plt.show()
```

```python
#Marr Hildreth edge detection
from skimage import filters

def marr_hildreth_edge_detection(img, sigma=1.0,threshold=0.01):
  log = filters.laplace(filters.gaussian(img,sigma))

  edges = np.zeros_like(log, dtype=np.uint8)
  for i in range(1, log.shape[0]-1):
        for j in range(1, log.shape[1]-1):
            patch = log[i-1:i+2, j-1:j+2]
            p_min, p_max = patch.min(), patch.max()
            if p_min < 0 and p_max > 0 and (p_max - p_min) > threshold:
                edges[i, j] = 255

    return edges

marr_hildreth_edges = marr_hildreth_edge_detection(img, 2, 0.01)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img, cmap="gray")
axes[0].set_title("Original Image")
axes[0].set_axis_off()
axes[1].imshow(marr_hildreth_edges, cmap="gray")
axes[1].set_title("Marr Hildreth Edge Detection")
axes[1].set_axis_off()
plt.tight_layout()
plt.show()
```

```python
#Marr Hildreth using custom function
from scipy.ndimage import convolve

def marr_hildreth(img, sigma=1.0, threshold=0.01):
    # 1. Gaussian smoothing
    blurred = cv2.GaussianBlur(img, (0, 0), sigma)

    # 2. Laplacian filtering
    laplacian = cv2.Laplacian(blurred, cv2.CV_64F)

    # 3. Zero-crossing detection
    edges = np.zeros_like(laplacian, dtype=np.uint8)
    for i in range(1, laplacian.shape[0]-1):
        for j in range(1, laplacian.shape[1]-1):
            patch = laplacian[i-1:i+2, j-1:j+2]
            p_min, p_max = patch.min(), patch.max()
            if p_min < 0 and p_max > 0 and (p_max - p_min) > threshold:
                edges[i, j] = 255

    return edges

marr_hildreth_edges_custom = marr_hildreth(img, 2, 5)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img, cmap="gray")
axes[0].set_title("Original Image")
axes[0].set_axis_off()
axes[1].imshow(marr_hildreth_edges_custom, cmap="gray")
axes[1].set_title("Marr Hildreth Edge Detection")
axes[1].set_axis_off()
plt.tight_layout()
plt.show()

```

```python
image_path = "./Lena.png"
img = cv2.imread(image_path)

if img is None:
    print("Failed to load image:", image_path)
    sys.exit(1)

max_width = 600
max_height = 600
h, w = img.shape[:2]
if w > max_width:
    scale = max_width / w
    new_w = max_width
    new_h = int(h * scale)
    img = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_AREA)

if h > max_height:
    scale = max_height / h
    new_h = max_height
    new_w = int(w * scale)
    img = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_AREA)

# Clone of original for reset
img_orig = img.copy()

# Globals for mouse callback
drawing = False         # True if mouse is pressed (for painting)
rect_drawing = False    # True if drawing rectangle (right button)
ix, iy = -1, -1         # initial positions
rect = (0,0,1,1)        # rectangle for init
rect_or_mask = 0        # 0 = draw rect, 1 = use mask
mask = np.zeros(img.shape[:2], dtype=np.uint8)  # mask initialized to PR_BG
# mask can have values: cv2.GC_BGD=0, cv2.GC_FGD=1, cv2.GC_PR_BGD=2, cv2.GC_PR_FGD=3
bgdModel = np.zeros((1,65), np.float64)
fgdModel = np.zeros((1,65), np.float64)

# Drawing brush settings
brush_radius = 5
value_fg = cv2.GC_FGD       # definite foreground
value_bg = cv2.GC_BGD       # definite background
drawing_mode = 'fg'        # 'fg' or 'bg'

# Colors for display (BGR)
COLOR_FG = (0,255,0)   # green for foreground strokes
COLOR_BG = (0,0,255)   # red for background strokes
COLOR_RECT = (255,0,0) # blue rect

# Canvas to show strokes on top of image
display = img.copy()

def reset_all():
    global img, display, mask, rect, rect_or_mask, bgdModel, fgdModel
    img = img_orig.copy()
    display = img.copy()
    mask = np.zeros(img.shape[:2], dtype=np.uint8)
    rect = (0,0,1,1)
    rect_or_mask = 0
    bgdModel = np.zeros((1,65), np.float64)
    fgdModel = np.zeros((1,65), np.float64)
    print("[INFO] Reset done.")

def apply_grabcut(iter_count=5):
    global mask, rect, rect_or_mask, bgdModel, fgdModel, img, display
    if rect_or_mask == 0:
        # Use rectangle initialization
        x,y,w,h = rect
        if w <= 0 or h <= 0:
            print("[WARN] Draw a rectangle first (right-drag) before applying GrabCut.")
            return
        print(f"[INFO] Running grabCut with rectangle {rect} and {iter_count} iterations.")
        cv2.grabCut(img, mask, (x,y,w,h), bgdModel, fgdModel, iter_count, cv2.GC_INIT_WITH_RECT)
        rect_or_mask = 1
    else:
        # Use mask refinement
        print(f"[INFO] Running grabCut with mask refinement and {iter_count} iterations.")
        cv2.grabCut(img, mask, None, bgdModel, fgdModel, iter_count, cv2.GC_INIT_WITH_MASK)

    # Build display: probable/definite foreground considered foreground
    mask2 = np.where((mask==cv2.GC_FGD) | (mask==cv2.GC_PR_FGD), 255, 0).astype('uint8')
    result = cv2.bitwise_and(img, img, mask=mask2)
    display = img.copy()
    # Blend result with original for visualization
    alpha = 0.6
    overlay = img.copy()
    overlay[mask2==0] = (overlay[mask2==0] * 0.4).astype(np.uint8)
    display = cv2.addWeighted(display, alpha, overlay, 1-alpha, 0)
    # Draw strokes on display for clarity: we'll draw definite fg/bg in top-left as legend
    cv2.putText(display, "Mode: " + ("FG" if drawing_mode=='fg' else "BG"), (10,20),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255,255,255), 1, cv2.LINE_AA)

def onmouse(event, x, y, flags, param):
    global ix, iy, drawing, rect_drawing, img, display, mask, rect, rect_or_mask

    # Right button: rectangle start/finish
    if event == cv2.EVENT_RBUTTONDOWN:
        rect_drawing = True
        ix, iy = x, y
    elif event == cv2.EVENT_RBUTTONUP:
        rect_drawing = False
        x0, y0 = min(ix, x), min(iy, y)
        x1, y1 = max(ix, x), max(iy, y)
        rect = (x0, y0, x1 - x0, y1 - y0)
        rect_or_mask = 0
        print(f"[INFO] Rectangle set to {rect}. Press 'n' to run GrabCut with this rectangle.")
        # Draw the rectangle on display
        display[:] = img.copy()
        cv2.rectangle(display, (x0,y0), (x1,y1), COLOR_RECT, 2)

    # Left button: paint strokes for fg/bg
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y
        paint_at(x, y)
    elif event == cv2.EVENT_MOUSEMOVE and drawing:
        paint_at(x, y)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        paint_at(x, y)

    # If currently drawing rectangle, show dynamic rectangle
    if rect_drawing:
        display[:] = img.copy()
        cv2.rectangle(display, (ix,iy), (x,y), COLOR_RECT, 2)

def paint_at(x, y):
    """Paint a circle on mask and display according to current drawing_mode."""
    global mask, display
    if drawing_mode == 'fg':
        # mark definite foreground
        cv2.circle(mask, (x,y), brush_radius, value_fg, -1)
        cv2.circle(display, (x,y), brush_radius, COLOR_FG, -1)
    else:
        # mark definite background
        cv2.circle(mask, (x,y), brush_radius, value_bg, -1)
        cv2.circle(display, (x,y), brush_radius, COLOR_BG, -1)

# Window and callback
win_name = "Interactive GraphCut (GrabCut)"
cv2.namedWindow(win_name)
cv2.setMouseCallback(win_name, onmouse)
display = img.copy()

print("""
Interactive GrabCut Keys:
    r - reset
    f - switch to foreground brush (draw with left mouse button)
    b - switch to background brush
    n - run GrabCut / refine segmentation
    s - save result to result.png
    + / - - increase / decrease brush size
    q / Esc - quit
Right-drag to draw initial rectangle before pressing 'n'.
""")

while True:
    # Show display
    disp_show = display.copy()
    # Draw legend of current brush size and mode
    cv2.putText(disp_show, f"Brush: {brush_radius}", (10, disp_show.shape[0]-10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255,255,255), 1, cv2.LINE_AA)
    cv2.imshow(win_name, disp_show)
    k = cv2.waitKey(20) & 0xFF

    if k == 27 or k == ord('q'):  # ESC / q
        break
    elif k == ord('r'):
        reset_all()
    elif k == ord('f'):
        drawing_mode = 'fg'
        print("[INFO] Mode: Foreground brush")
    elif k == ord('b'):
        drawing_mode = 'bg'
        print("[INFO] Mode: Background brush")
    elif k == ord('n'):
        apply_grabcut(iter_count=5)
    elif k == ord('s'):
        # Save current segmentation result
        mask2 = np.where((mask==cv2.GC_FGD) | (mask==cv2.GC_PR_FGD), 255, 0).astype('uint8')
        result = cv2.bitwise_and(img_orig, img_orig, mask=mask2)
        cv2.imwrite('result.png', result)
        print("[INFO] Saved result to result.png")
    elif k == ord('+') or k == ord('='):
        brush_radius = min(100, brush_radius + 1)
        print(f"[INFO] brush_radius = {brush_radius}")
    elif k == ord('-') or k == ord('_'):
        brush_radius = max(1, brush_radius - 1)
        print(f"[INFO] brush_radius = {brush_radius}")

cv2.destroyAllWindows()
```

#Mean shift segmentation


```python
img = cv2.imread("img")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

spatial_radius = 20 #How far the algorithm should look for at pixels for averaging
color_radius = 40 #How much color difference is allowed while grouping pixels
max_pyramid_level = 2 #How much downsample to use for faster computation

mean_shift = v2.pyrMeanShiftFiltering(img,spatial_radius,color_radius,maxpyramid_level)

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Mean shift segmented image")
plt.imshow(mean_shift)
plt.axis("off")
plt.show()
```

```python
def gaussian_kernel(distance_sq, bandwidth):
    return np.exp(-0.5 * distance_sq / (bandwidth ** 2))

def mean_shift_segmentation(X, bandwidth=20, max_iter=10, tol=1e-3):
    X = np.array(X, dtype=float)
    n_samples, n_features = X.shape
    shifted_points = np.copy(X)

    print("Inside Segmentation function")

    for it in range(max_iter):
        # Compute pairwise distances (vectorized, memory heavy)
        distances = np.sum((shifted_points[:, None, :] - X[None, :, :]) ** 2, axis=2)
        weights = gaussian_kernel(distances, bandwidth)  # (n_samples, n_samples)
        numerator = weights @ X  # weighted sum
        denominator = np.sum(weights, axis=1)[:, None]
        new_points = numerator / denominator

        shift_distances = np.linalg.norm(new_points - shifted_points, axis=1)
        max_shift = np.max(shift_distances)
        shifted_points = new_points

        if max_shift < tol:
            print(f"Converged after {it+1} iterations")
            break

        print(f"Iteration:{it+1}/{max_iter}")



    # Merge nearby cluster centers
    cluster_centers = []
    labels = np.zeros(n_samples)
    cluster_tol = bandwidth / 2

    for i, point in enumerate(shifted_points):
        found_cluster = False
        for j, center in enumerate(cluster_centers):
            if np.linalg.norm(point - center) < cluster_tol:
                labels[i] = j
                found_cluster = True
                break
        if not found_cluster:
            cluster_centers.append(point)
            labels[i] = len(cluster_centers) - 1

    return np.array(cluster_centers), labels


# ---- Image Segmentation ----
img = cv2.imread("img")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
h, w, c = img.shape

# Parameters (important!)
spatial_radius = 12
color_radius = 32
bandwidth = 6
max_iter = 15

X = []
for i in range(0, h, 2):
    for j in range(0, w, 2):
        R, G, B = img[i, j]
        X.append([i / spatial_radius, j / spatial_radius,
                  R / color_radius, G / color_radius, B / color_radius])
X = np.array(X)

centers, labels = mean_shift_segmentation(X, bandwidth=bandwidth, max_iter=max_iter)

segmented = centers[labels.astype(int), 2:5] * color_radius
segmented = segmented.reshape(len(range(0, h, 2)), len(range(0, w, 2)), 3).astype(np.uint8)

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Mean Shift Segmentation")
plt.imshow(segmented)
plt.axis("off")

plt.show()
```

1. Images where the objects close to each other have similar color variation(color gradient) will be segemented as one object.
2. Required manual tuning of the paramters for getting proper segmentation.
3. Semi supervised in terms of choosing the parameters i.e. spatial and color radius.

#SIFT Implementation

```python
label_map = pd.read_csv("names.csv")  # columns: label, name
label_dict = dict(zip(label_map["label"], label_map["name"]))

sift = cv2.SIFT_create()
train_features = {}
train_labels = []

train_path = "photos"

print("Extracting SIFT features from training images...")

for label, person_name in label_dict.items():
    person_folder = os.path.join(train_path, str(label))
    if not os.path.exists(person_folder):
        print(f"Folder missing for {person_name}")
        continue

    descriptors_list = []
    for img_name in os.listdir(person_folder):
        img_path = os.path.join(person_folder, img_name)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            continue

        keypoints, descriptors = sift.detectAndCompute(img, None)
        if descriptors is not None:
            descriptors_list.append(descriptors)

    if descriptors_list:
        # Combine all descriptors for this person
        train_features[label] = np.vstack(descriptors_list)
        train_labels.append(label)

print("Training feature extraction complete.\n")
```

```python
def match_sift_features(desc1,desc2):
  bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck= True)
  matches = bf.match(desc1,desc2)
  matches = sorted(matches, key = lambda x:x.distance)
  return matches
```

```python
test_csv = pd.read_csv("test.csv")  # columns: image,label
test_path = "test_photos"

correct = 0
total = 0

print("Matching test images with training data...\n")

for idx, row in test_csv.iterrows():
    test_img_name = row["file_name"]
    true_label = row["label"]

    test_img_path = os.path.join(test_path, f"{test_img_name}.jpeg")
    test_img = cv2.imread(test_img_path, cv2.IMREAD_GRAYSCALE)
    if test_img is None:
        print(f" Could not read {test_img_name}")
        continue

    kp_test, desc_test = sift.detectAndCompute(test_img, None)
    if desc_test is None:
        print(f" No SIFT features found for {test_img_name}")
        continue

    best_label = None
    best_score = float("inf")

    # Compare with all training persons
    for label, train_desc in train_features.items():
        matches = match_sift_features(desc_test, train_desc)

        # Take mean distance of top 50 matches
        num_matches = min(50, len(matches))
        mean_distance = np.mean([m.distance for m in matches[:num_matches]])

        if mean_distance < best_score:
            best_score = mean_distance
            best_label = label

    predicted_name = label_dict[best_label]
    true_name = label_dict[true_label]

    print(f"Test: {test_img_name:20s} | Predicted: {predicted_name:10s} | Actual: {true_name}")

    if best_label == true_label:
        correct += 1
    total += 1
```

```python
accuracy = (correct / total) * 100 if total > 0 else 0
print(f"\nAccuracy: {accuracy:.2f}% ({correct}/{total} correct)")
```
